# Backtracking

## N-Queens

```python
board = [4, 7, 2, 6, 1, 0, 3, 5]

for row, col in enumerate(board):
  print(f"Hay una reina en la fila {row} columna {col}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw(board):
  n = len(board)
  b = np.zeros((n, n, 3), dtype=int)
  b += [255, 128, 80]

  b[::2, ::2] = [255, 225, 120]
  b[1::2, 1::2] = [255, 225, 120]

  _, ax = plt.subplots()
  ax.imshow(b)
  for row, col in enumerate(board):
    ax.text(col, row, u"\u265b", fontsize=200/n, va="center", ha="center")
  ax.set(xticks=[], yticks=[])

draw(board)
```

```python
def valid(board, row, col):
  for row_i in range(row):
    col_i = board[row_i]
    delta = row - row_i
    if col in [col_i, col_i + delta, col_i - delta]:
      return False

  return True
```

```python
board = [1, 3, -1, -1]
assert valid(board, 2, 0) == True
draw(board)
```

```python
board = [1, 3, -1, -1]
assert valid(board, 2, 1) == False
draw(board)
```

```python
board = [1, 3, -1, -1]
assert valid(board, 2, 2) == False
draw(board)
```

```python
def nqueens(board, row):
  n = len(board)
  if row == n:
    draw(board)
  else:
    for col in range(n):
      if valid(board, row, col):
        board[row] = col
        nqueens(board, row+1)
```
