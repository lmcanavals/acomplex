```python
def bfs_am(G, s):
  n = len(G)
  visited = [False]*n
  path = [-1]*n # parent
  queue = [s]
  visited[s] = True

  while queue:
    u = queue.pop(0)
    for v in range(n):
      if G[u, v] == 1 and not visited[v]:
        visited[v] = True
        path[v] = u
        queue.append(v)

  return path
```

```python
import numpy as np
import graphviz as gv
```

```python
G1 = np.loadtxt("0403a.am", dtype=int)
```

```python
def drawG_am(G, directed=False, path=None):
  if directed:
    graph = gv.Digraph("felicidad")
  else:
    graph = gv.Graph("alegria")
  n = len(G)
  for i in range(n):
    for j in range(0 if directed else i, n):
      if G[i, j] == 1:
        if path != None and (path[i] == j or path[j] == i):
          graph.edge(str(i), str(j), color="orange")
        else:
          graph.edge(str(i), str(j))
  return graph
```

```python
drawG_am(G1)
```
